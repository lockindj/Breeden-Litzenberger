```python
# ============================================================
# Simple yearly snapshots + realized (+30 next trading day) all in pandas
# - No extra SQL complexity (mirrors your working puts query style)
# - Build the +30 mapping and realized underlying in pandas
# ============================================================
import pandas as pd
import numpy as np

# ---------- 1) Get min/max from each table (split queries to avoid timeout) ----------
date_range_calls = SQL_dataframe("""
    SELECT MIN(quote_date) AS min_date, MAX(quote_date) AS max_date
    FROM calls_table;
""")
min_calls = date_range_calls['min_date'][0]
max_calls = date_range_calls['max_date'][0]

date_range_puts = SQL_dataframe("""
    SELECT MIN(quote_date) AS min_date, MAX(quote_date) AS max_date
    FROM puts_table;
""")
min_puts = date_range_puts['min_date'][0]
max_puts = date_range_puts['max_date'][0]

print("calls_table range:", min_calls, "→", max_calls)
print("puts_table  range:", min_puts,  "→", max_puts)

# ---------- 2) Yearly batching (like your working query) ----------
def fetch_snapshots_calls(min_date, max_date):
    batch_starts = pd.date_range(start=min_date, end=max_date, freq='YS')
    results = []
    for start in batch_starts:
        end = (start + pd.DateOffset(years=1)) - pd.DateOffset(days=1)
        q = f"""
            SELECT 
                id,
                quote_date,
                underlying_last,
                dte,
                strike,
                strike / underlying_last AS moneyness,
                c_bid,
                c_ask,
                c_last,
                expire_date
            FROM calls_table
            WHERE dte = 30
              AND c_bid IS NOT NULL
              AND c_ask IS NOT NULL
              AND c_iv  IS NOT NULL
              AND underlying_last IS NOT NULL
              AND underlying_last > 0
              AND quote_date BETWEEN '{start.date()}' AND '{end.date()}'
        """
        df = SQL_dataframe(q)
        results.append(df)
    return pd.concat(results, ignore_index=True) if results else pd.DataFrame()

def fetch_snapshots_puts(min_date, max_date):
    batch_starts = pd.date_range(start=min_date, end=max_date, freq='YS')
    results = []
    for start in batch_starts:
        end = (start + pd.DateOffset(years=1)) - pd.DateOffset(days=1)
        q = f"""
            SELECT 
                id,
                quote_date,
                underlying_last,
                dte,
                strike,
                strike / underlying_last AS moneyness,
                p_bid,
                p_ask,
                p_last,
                expire_date
            FROM puts_table
            WHERE dte = 30
              AND p_bid IS NOT NULL
              AND p_ask IS NOT NULL
              AND p_iv  IS NOT NULL
              AND underlying_last IS NOT NULL
              AND underlying_last > 0
              AND quote_date BETWEEN '{start.date()}' AND '{end.date()}'
        """
        df = SQL_dataframe(q)
        results.append(df)
    return pd.concat(results, ignore_index=True) if results else pd.DataFrame()

df_snapshots_calls = fetch_snapshots_calls(min_calls, max_calls)
df_snapshots_puts  = fetch_snapshots_puts(min_puts,  max_puts)

print("Calls snapshots:", df_snapshots_calls.shape)
print("Unique call quote_dates:", df_snapshots_calls["quote_date"].nunique() if not df_snapshots_calls.empty else 0)
print("Puts snapshots:", df_snapshots_puts.shape)
print("Unique put quote_dates:", df_snapshots_puts["quote_date"].nunique() if not df_snapshots_puts.empty else 0)

# ---------- 3) Build +30 next-trading-day mapping in pandas ----------
def attach_realized_from_self(df_snapshots, quote_col="quote_date", ul_col="underlying_last"):
    """
    For a snapshot DataFrame from a single table:
    - Build calendar of available trading days from `quote_col`
    - For each quote_date, map to the next available trading day >= quote_date + 30d
    - Join realized_underlying from per-day underlying_last (groupby max)
    Returns df with added columns: realized_date, realized_underlying
    """
    if df_snapshots.empty:
        df_snapshots[["realized_date", "realized_underlying"]] = np.nan
        return df_snapshots

    # Ensure datetime
    df = df_snapshots.copy()
    df[quote_col] = pd.to_datetime(df[quote_col])

    # Unique trading calendar from the snapshot
    cal = (df[[quote_col]]
           .drop_duplicates()
           .sort_values(quote_col)
           .rename(columns={quote_col: "cal_date"})
           .reset_index(drop=True))

    # Target = quote_date + 30 calendar days
    key = df[[quote_col]].drop_duplicates().sort_values(quote_col).reset_index(drop=True)
    key["target"] = key[quote_col] + pd.Timedelta(days=30)

    # Map to next available trading day using merge_asof
    mapped = pd.merge_asof(
        key.sort_values("target"),
        cal.sort_values("cal_date"),
        left_on="target",
        right_on="cal_date",
        direction="forward",
        allow_exact_matches=True
    ).rename(columns={"cal_date": "realized_date"})[[quote_col, "realized_date"]]

    # Per-day underlying_last (max is safe if constant intraday)
    ul_by_day = (df.groupby(quote_col, as_index=False)[ul_col]
                   .max()
                   .rename(columns={quote_col: "realized_date",
                                    ul_col: "realized_underlying"}))

    # Merge realized_date into snapshots, then realized_underlying
    df = df.merge(mapped, on=quote_col, how="left")
    df = df.merge(ul_by_day, on="realized_date", how="left")

    return df

df_snapshots_calls = attach_realized_from_self(df_snapshots_calls,
                                              quote_col="quote_date",
                                              ul_col="underlying_last")

df_snapshots_puts  = attach_realized_from_self(df_snapshots_puts,
                                              quote_col="quote_date",
                                              ul_col="underlying_last")

# ---------- 4) Final checks ----------
print("Calls (with realized):", df_snapshots_calls.shape,
      "| missing realized:", df_snapshots_calls["realized_underlying"].isna().sum())

print("Puts  (with realized):", df_snapshots_puts.shape,
      "| missing realized:", df_snapshots_puts["realized_underlying"].isna().sum())

# Optionally drop rows without a realized trading day found
df_snapshots_calls = df_snapshots_calls.dropna(subset=["realized_date","realized_underlying"])
df_snapshots_puts  = df_snapshots_puts.dropna(subset=["realized_date","realized_underlying"])

print("Calls (after dropna):", df_snapshots_calls.shape)
print("Puts  (after dropna):", df_snapshots_puts.shape)
```

# Put–Call Parity (European, 30-DTE): Definition, Intuition, and Use in Our Pipeline

## 1) Parity identity (European options)
For European options on spot $S_0$ with risk-free rate $r$, continuous dividend yield $q$, maturity $T$ (here fixed at 30 days), and strike $K$:
$C(K,T) - P(K,T) = S_0 e^{-qT} - K e^{-rT}.$

Define the discount factor $D = e^{-rT}$ and the forward $F = S_0 e^{(r-q)T}$. Then:
$C - P = D\,(F - K), \quad C = P + D(F-K), \quad P = C - D(F-K).$

**Assumptions:** European exercise, same $(K,T)$, frictionless markets; small deviations within bid–ask are not exploitable arbitrage.

---

## 2) Economic intuition
The payoff of **long call + short put** with the same $(K,T)$ at expiry is
$(S_T-K)^+ - (K-S_T)^+ = S_T - K$ — i.e., a **long forward** struck at $K$.
Its present value is $S_0 e^{-qT} - K e^{-rT}$, hence the parity identity $C-P = S_0 e^{-qT} - K e^{-rT}$.

---

## 3) Why we use parity here
### 3.1 Unifying calls and puts into one coherent cross-section
We want a single, clean call-price curve across strikes to fit SVI and to feed Breeden–Litzenberger (BL). Parity lets us anchor each wing with the more reliable instrument:

- We use your **day-specific** $r$ and $q$ to compute $F = S_0 e^{(r-q)T}$ and $D = e^{-rT}$.
- For **$K/F < 1$** (left wing): prefer **OTM puts**; build **synthetic calls** via $C_{\text{synth}}(K)=P_{\text{mid}}(K)+D(F-K)$.
- For **$K/F \ge 1$** (right wing): prefer **actual calls** $C_{\text{mid}}(K)$.
- If only one side exists at $K$, use it; if both exist, we use mid prices and rely on robust fitting downstream (no spread-based weighting).

This reduces dependence on noisy deep ITM calls on the left and yields a parity-consistent smile across $K$.

### 3.2 Consistency check (optional)
Where both $C_{\text{mid}}$ and $P_{\text{mid}}$ exist, a simple diagnostic is the **parity residual**
$R(K)=C_{\text{mid}}(K) - \big(P_{\text{mid}}(K) + D(F-K)\big)$,
which should be small and centered near $0$ (within a tolerance comparable to typical half-spreads). We **do not** use spreads for weights; we keep mid prices and a robust loss in calibration.

---

## 4) How parity plugs into SVI → BL

### 4.1 Data preparation (per day, 30-DTE)
- Use day-specific $S_0$, $r$, $q$, and $T$ (30/365).
- Compute $F$ and $D$.
- Build a **unified call curve** $C(K)$:
  - If $K/F<1$: use $C_{\text{synth}}(K)$ if a put mid exists; otherwise use $C_{\text{mid}}(K)$.
  - If $K/F\ge 1$: use $C_{\text{mid}}(K)$ if available; otherwise use $C_{\text{synth}}(K)$.
  - Keep one call price per strike and sort by $K$.

### 4.2 SVI calibration (moneyness $K/F$)
- **Calibration band (fit):** use observed strikes with $K/F \in [0.85,\,1.15]$ (adjustable).
- Fit SVI in total-variance space with a **robust loss** (e.g., soft-L1).
- Verify **in-band** no-arb: call price decreases in $K$ (monotonicity) and is convex in $K$ (convexity); check **Lee’s wing-slope bounds**.

### 4.3 Evaluation and BL density
- **Evaluation band:** can be wider than the calibration band (e.g., $[0.75,\,1.30]$) to let BL “see” tails; we also report how much of this band lies **outside observed strikes** (extrapolation diagnostics).
- Price a **uniform strike grid** (e.g., 1201 points) with the fitted SVI.
- Compute the BL density by numerical differentiation:
  $q(K)=e^{rT}\,\frac{\partial^2 C(K)}{\partial K^2}$ (central differences; clip tiny negatives to $0$).
- Check the **total probability mass** on the band
  $\int q(K)\,dK = \texttt{total\_eval\_mass}$ and report **left/right tail mass** outside observed coverage.

### 4.4 Quality gate
- **Filter** days with $|\texttt{total\_eval\_mass}-1|>0.001$ (0.1%) per your criterion.
- For analysis/score computation you may **renormalize** $q$ to integrate to $1$, while still logging the raw mass and tail fractions.

---

## 5) Limitations and scope
- **European options:** Parity holds exactly (your dataset complies).
- **Dividends:** Using day-specific $q$ ensures $F=S_0 e^{(r-q)T}$ is consistent.
- **Mids only:** We use mid prices plus robust calibration; spread data are used only as a diagnostic tolerance, not as weights.

---

## 6) Summary — what parity buys us
- A **single, coherent** call-price cross-section across strikes, anchored by the more reliable instrument on each wing.
- **Internal consistency** between calls and puts via $C-P = D(F-K)$.
- **More stable SVI slices** and **more credible BL left tails**, with transparent diagnostics of how much probability mass comes from extrapolated regions.

```python
# === USE df_enriched -> unified calls (parity with external r,q) -> IVs (your function) ===
import numpy as np
import pandas as pd
from math import exp, isfinite

# --- helpers ---
def _to_bool_is_call(series):
    if series.dtype == bool or str(series.dtype) == "boolean":
        return series.astype(bool)
    def cast(x):
        if pd.isna(x): return np.nan
        if isinstance(x, (bool, np.bool_)): return bool(x)
        if isinstance(x, (int, np.integer, float, np.floating)): return bool(int(x))
        if isinstance(x, str):
            t = x.strip().upper()
            if t in ("C","CALL","1","T","TRUE"): return True
            if t in ("P","PUT","0","F","FALSE"): return False
        return np.nan
    return series.apply(cast).astype("boolean").astype(bool)

def _mid_from_bid_ask_last(df, bid_col="bid", ask_col="ask", last_col="last"):
    has_ba = df[[bid_col, ask_col]].notna().all(axis=1)
    return np.where(
        has_ba,
        (pd.to_numeric(df[bid_col], errors="coerce") + pd.to_numeric(df[ask_col], errors="coerce")) / 2.0,
        pd.to_numeric(df.get(last_col, np.nan), errors="coerce"),
    )

def _unified_calls_for_day_external(df_day: pd.DataFrame, debug=False) -> pd.DataFrame:
    """
    Build parity-consistent unified call curve for one day using external r,q in df_day.
    Returns rows with columns expected by your compute_iv_on_dataframe:
      ['quote_date','underlying_last','dte_days','strike',
       'risk_free_rate','Annualized_Dividend_Yield','is_call','mid_price']
    """
    # Coerce numerics
    num_cols = ["underlying_last","dte","strike","bid","ask","last","div_yield_annual","risk_free_1m"]
    for c in num_cols:
        if c in df_day.columns:
            df_day[c] = pd.to_numeric(df_day[c], errors="coerce")

    df_day = df_day.dropna(subset=["quote_date","underlying_last","dte","strike"])
    if df_day.empty:
        if debug: print("  -> empty after basic clean (missing quote_date/underlying_last/dte/strike)")
        return pd.DataFrame()

    qdate    = pd.to_datetime(df_day["quote_date"].iloc[0]).date()
    S        = float(df_day["underlying_last"].iloc[0])
    dte_days = float(np.median(df_day["dte"]))
    T_yrs    = dte_days / 365.0
    r        = float(np.median(pd.to_numeric(df_day["risk_free_1m"], errors="coerce").dropna()))
    q        = float(np.median(pd.to_numeric(df_day["div_yield_annual"], errors="coerce").dropna()))

    if not (isfinite(S) and isfinite(T_yrs) and T_yrs > 0 and isfinite(r) and isfinite(q)):
        if debug: print(f"  -> bad snapshot S/T/r/q: S={S}, T_yrs={T_yrs}, r={r}, q={q}")
        return pd.DataFrame()

    # Ensure r,q are decimals (not %). If they’re >1.0 we interpret as percent.
    if r > 1.0: r = r / 100.0
    if q > 1.0: q = q / 100.0

    D = exp(-r*T_yrs)
    F = S * exp((r - q) * T_yrs)

    # Split sides and compute mids
    df_day["is_call"] = _to_bool_is_call(df_day["is_call"])
    calls = df_day.loc[df_day["is_call"] == True].copy()
    puts  = df_day.loc[df_day["is_call"] == False].copy()

    calls["c_mid"] = _mid_from_bid_ask_last(calls, "bid", "ask", "last")
    puts["p_mid"]  = _mid_from_bid_ask_last(puts,  "bid", "ask", "last")

    # Basic diagnostics
    if debug:
        print(f"  S={S:.4f}, T={T_yrs:.5f}y, r={r:.6f}, q={q:.6f}, F={F:.4f}")
        print(f"  counts: calls={len(calls)}, puts={len(puts)}")
        print(f"  mids>0: c_mid={(np.nan_to_num(calls['c_mid'])>0).sum()}, p_mid={(np.nan_to_num(puts['p_mid'])>0).sum()}")

    # Build grid (outer join on strikes)
    grid = pd.merge(
        calls[["strike","c_mid"]],
        puts[["strike","p_mid"]],
        on="strike", how="outer"
    ).sort_values("strike").reset_index(drop=True)

    # Synthetic call from put via parity: C = P + D*(F - K)
    grid["c_synth"] = np.where(grid["p_mid"].notna(), grid["p_mid"] + D*(F - grid["strike"]), np.nan)

    # Preference rule + fallbacks
    cond_right = grid["strike"] >= F
    c_best = np.where(cond_right, grid["c_mid"], grid["c_synth"])
    c_best = np.where(np.isnan(c_best), grid["c_mid"], c_best)     # fallback to actual call
    c_best = np.where(np.isnan(c_best), grid["c_synth"], c_best)   # fallback to synthetic
    grid["mid_price"] = c_best

    out = grid.dropna(subset=["mid_price"]).copy()
    # Allow zero-priced options? Usually 0 mids are unusable; keep > 0 to avoid IV failures
    out = out.loc[out["mid_price"] > 0]

    if debug:
        print(f"  unified rows: {len(out)} (min/max strike: {out['strike'].min() if not out.empty else np.nan} / {out['strike'].max() if not out.empty else np.nan})")

    if out.empty:
        return pd.DataFrame()

    # Final columns expected by your IV function
    out["quote_date"]                = qdate
    out["underlying_last"]           = S
    out["dte_days"]                  = dte_days
    out["risk_free_rate"]            = r
    out["Annualized_Dividend_Yield"] = q
    out["is_call"]                   = True  # unified curve is calls-only

    return out[[
        "quote_date","underlying_last","dte_days","strike",
        "risk_free_rate","Annualized_Dividend_Yield","is_call","mid_price"
    ]]

def build_unified_calls_dataframe_external(df_enriched: pd.DataFrame, verbose: bool=True, debug_first_n: int = 3) -> pd.DataFrame:
    df = df_enriched.copy()
    df["quote_date"] = pd.to_datetime(df["quote_date"]).dt.date
    dates = pd.Series(df["quote_date"]).sort_values().unique()

    frames, skipped = [], []
    for i, d in enumerate(dates):
        df_day = df.loc[df["quote_date"] == d].copy()
        part = _unified_calls_for_day_external(df_day, debug=(i < debug_first_n))
        if not part.empty:
            frames.append(part)
        else:
            # Collect a reason summary
            n_calls = int((df_day["is_call"]==True).sum()) if "is_call" in df_day else -1
            n_puts  = int((df_day["is_call"]==False).sum()) if "is_call" in df_day else -1
            reason = f"no_usable_unified_calls (calls={n_calls}, puts={n_puts})"
            skipped.append((d, reason))

    out = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame(columns=[
        "quote_date","underlying_last","dte_days","strike",
        "risk_free_rate","Annualized_Dividend_Yield","is_call","mid_price"
    ])
    if verbose:
        print(f"Unified call rows: {len(out)} across {out['quote_date'].nunique()} days. Skipped {len(skipped)} days.")
        if skipped[:5]:
            print("First few skips:", skipped[:5])
    return out

# ---- RUN on df_enriched (make sure you're passing THIS enriched frame) ----
# Your IV function expects these column names:
column_map = {
    "S":"underlying_last",
    "K":"strike",
    "DTE":"dte_days",
    "r":"risk_free_rate",
    "q":"Annualized_Dividend_Yield",
    "is_call":"is_call",
    "price":"mid_price",
}

# IMPORTANT: use df_enriched (the one you just merged r & q into)
df_unified_calls = build_unified_calls_dataframe_external(df_enriched, verbose=True, debug_first_n=2)

# Now compute IVs with your function (must already be imported/defined)
df_iv = compute_iv_on_dataframe(
    df_unified_calls,
    column_map=column_map,
    dte_unit="days",
    rates_in_percent=False,  # r,q are decimals now
    n_jobs=1,
)

print("IV rows:", len(df_iv))
display(df_unified_calls.head(5))
display(df_iv.head(5))
```

```python
# ================================================================
# Full SVI + BL pipeline (with detailed comments)
# ------------------------------------------------
# What this cell provides:
#   • A robust SVI slice fitter (`fit_svi`) working from your df_iv
#   • Pricing from SVI on a strike grid (`price_from_svi`)
#   • No-arbitrage checks in price space (`convexity_report`)
#   • BL density extraction and tail-mass diagnostics
#   • A driver to run this over all days with a dynamic evaluation band
#
# Assumptions about your input `df_iv`:
#   df_iv has at least the following columns:
#     - quote_date           (datetime-like or 'YYYY-MM-DD' string)
#     - underlying_last      (spot S, scalar per day)
#     - risk_free_1m         (risk-free rate r, *decimal* per year)
#     - div_yield_annual     (dividend yield q, *decimal* per year)
#     - dte                  (time to expiry in DAYS; we convert to years)
#     - strike               (option strikes)
#     - implied_volatility   (per-strike IV *decimal*, e.g. 0.20)
#
# Output:
#   - summary_df : one row per day with diagnostics, tail mass metrics, etc.
#   - artifacts  : dict[quote_date] -> curves & reports for deeper inspection
#
# Notes:
#   • Calibration band is defined in moneyness K/S for data selection.
#   • SVI is fitted in total variance w(k) with k = ln(K/F) (forward log-moneyness).
#   • Evaluation band is dynamic per day: observed moneyness ± allowed_gap (clamped).
#   • BL density q(K) = e^{rT} * d^2C/dK^2 via central differences.
# ================================================================

import numpy as np
import pandas as pd
from math import log, exp, sqrt, isfinite
from scipy.optimize import least_squares
from scipy.stats import norm

# ──────────────────────────────────────────────────────────────────────────────
# Black–Scholes pricing (fallback) + vectorized call helper
# If you already have bs_price() imported from your own library, this fallback
# will be ignored if you comment it out.
# ──────────────────────────────────────────────────────────────────────────────
def _bs_price_fallback(S, K, T, r, q, sig, is_call=True):
    """
    European Black–Scholes with continuous dividend yield q.
    S, K > 0; T >= 0; r, q real; sig > 0.
    """
    if T <= 0:
        # At expiry, option price = intrinsic (discounting dividend not needed at T=0)
        if is_call:
            return max(S - K, 0.0)
        else:
            return max(K - S, 0.0)
    sig = max(sig, 1e-12)
    d1 = (log(S / K) + (r - q + 0.5 * sig * sig) * T) / (sig * sqrt(T))
    d2 = d1 - sig * sqrt(T)
    DF  = exp(-q * T)  # dividend discount
    D   = exp(-r * T)  # risk-free discount
    if is_call:
        return DF * S * norm.cdf(d1) - D * K * norm.cdf(d2)
    else:
        return D * K * norm.cdf(-d2) - DF * S * norm.cdf(-d1)

# If you prefer to force using your own bs_price, replace the next line:
bs_price = _bs_price_fallback

# Vectorized call pricer for convenience
bs_call_vec = np.vectorize(lambda S, K, T, r, q, sig: bs_price(S, K, T, r, q, sig, True))


# ──────────────────────────────────────────────────────────────────────────────
# SVI (Stochastic Volatility Inspired) total variance parameterization
# w(k) = a + b * ( rho*(k - m) + sqrt( (k - m)^2 + sigma^2 ) )
#   where k = ln(K/F) is forward log-moneyness
# Parameters constraints for no-arbitrage wings include:
#   b >= 0, |rho| < 1, sigma > 0, and Lee wing-slope bounds b*(1±rho) <= 2
# ──────────────────────────────────────────────────────────────────────────────
def svi_w(k, a, b, rho, m, sigma):
    sigma = np.maximum(sigma, 1e-10)  # numerical guard
    return a + b * (rho * (k - m) + np.sqrt((k - m) ** 2 + sigma ** 2))


def lee_wing_slopes_ok(b, rho, tol=1e-8):
    """
    Lee's moment formula implies:
      left wing slope  = b*(1 - rho) <= 2
      right wing slope = b*(1 + rho) <= 2
    Returns (ok: bool, left_slope, right_slope).
    """
    left  = b * (1.0 - rho)
    right = b * (1.0 + rho)
    ok = (left <= 2.0 + tol) and (right <= 2.0 + tol) and (b >= 0.0)
    return ok, float(left), float(right)


# ──────────────────────────────────────────────────────────────────────────────
# Core SVI fit for a single day
# ──────────────────────────────────────────────────────────────────────────────
def fit_svi(
    df_IV: pd.DataFrame,
    target_date,
    calib_band=(0.80, 1.15),     # select observed strikes with K/S in this band for fitting
    robust=True,
    iv_clip=(1e-4, 5.0),         # clamp IVs to a sane numeric range before building total variance
):
    """
    Fit one SVI slice on a given day from df_IV rows with that quote_date.
    Returns a dict containing inputs, fitted params, and basic diagnostics.

    Steps:
      • Extract day slice and coerce numeric.
      • Build forward F = S*exp((r-q)T), time in years T = dte/365.
      • Select calibration points using K/S in [calib_band].
      • Fit SVI in total variance space w = (IV^2) * T via robust least squares (optional).
      • Return fitted parameters and convenient observables.
    """
    # Slice by date
    d = df_IV.loc[pd.to_datetime(df_IV["quote_date"]).dt.date == pd.to_datetime(target_date).date()].copy()
    if d.empty:
        raise ValueError(f"No rows for quote_date={target_date}")

    # Coerce numerics
    for c in ["strike", "implied_volatility", "underlying_last", "div_yield_annual", "risk_free_1m", "dte"]:
        if c in d.columns:
            d[c] = pd.to_numeric(d[c], errors="coerce")

    # Drop rows missing key fields
    d = d.dropna(subset=["strike", "implied_volatility", "underlying_last", "div_yield_annual", "risk_free_1m", "dte"])
    if d.empty:
        raise ValueError(f"All rows NA after cleaning for {target_date}")

    # Use a consistent snapshot per day
    S = float(d["underlying_last"].iloc[0])
    r = float(d["risk_free_1m"].median())
    q = float(d["div_yield_annual"].median())
    T = float(d["dte"].median()) / 365.0
    if not (isfinite(S) and isfinite(r) and isfinite(q) and isfinite(T) and T > 0):
        raise ValueError(f"Bad snapshot S/r/q/T for {target_date}")

    # Forward & strikes
    F = S * exp((r - q) * T)
    K = d["strike"].to_numpy(float)

    # Optional: clamp absurd IVs a bit, then build total variance
    iv = np.clip(d["implied_volatility"].to_numpy(float), iv_clip[0], iv_clip[1])
    w_obs = np.maximum(iv, 1e-10) ** 2 * T

    # Calibration set: select points by moneyness K/S ∈ calib_band
    mny_S = K / S
    sel = (mny_S >= calib_band[0]) & (mny_S <= calib_band[1])
    if sel.sum() < 5:
        # If too few points in the band, fall back to using all points (last resort)
        sel = np.isfinite(K) & np.isfinite(w_obs)

    K_cal = K[sel]
    w_cal = w_obs[sel]

    # Build forward log-moneyness for fitting
    k_cal = np.log(K_cal / F)

    # Parameter initialization (simple, robust heuristics)
    a0 = np.clip(np.percentile(w_cal, 10), 1e-6, None)
    span_k = np.percentile(k_cal, 90) - np.percentile(k_cal, 10) + 1e-8
    b0 = np.clip((np.percentile(w_cal, 90) - np.percentile(w_cal, 10)) / span_k, 1e-6, 10.0)
    rho0, m0 = 0.0, 0.0
    sigma0 = np.clip((k_cal.max() - k_cal.min()) / 4.0 if np.isfinite(k_cal.max() - k_cal.min()) else 0.2, 0.05, 2.0)
    x0 = np.array([a0, b0, rho0, m0, sigma0], float)

    # Parameter bounds (soft Lee constraints enter via post-check)
    lb = np.array([0.0, 1e-9, -0.999, -np.inf, 1e-6], float)
    ub = np.array([np.inf, np.inf,  0.999,  np.inf,  np.inf], float)

    # Residuals in total variance space
    def resid(x):
        a, b, rho, m, sigma = x
        wf = np.maximum(svi_w(k_cal, a, b, rho, m, sigma), 1e-12)
        return wf - w_cal

    # Least squares with optional robust loss
    kw = dict(x0=x0, bounds=(lb, ub), jac="2-point", max_nfev=20000)
    if robust:
        fscale = np.median(np.abs(w_cal - np.median(w_cal))) + 1e-6
        res = least_squares(resid, loss="soft_l1", f_scale=fscale, **kw)
    else:
        res = least_squares(resid, **kw)

    a, b, rho, m, sigma = res.x
    rmse_w = float(np.sqrt(np.mean(resid(res.x) ** 2)))

    lee_ok, slope_left, slope_right = lee_wing_slopes_ok(b, rho)

    # Pack result
    out = {
        "S": S, "r": r, "q": q, "T": T, "F": F,
        "K_obs": K, "iv_obs": iv, "k_obs": np.log(K / F),
        "mnyS_obs": mny_S,
        "params": {"a": float(a), "b": float(b), "rho": float(rho), "m": float(m), "sigma": float(sigma)},
        "rmse_w": rmse_w,
        "lee_ok": bool(lee_ok),
        "slope_left": float(slope_left),
        "slope_right": float(slope_right),
        "res_status": int(res.status)
    }
    return out


# ──────────────────────────────────────────────────────────────────────────────
# Pricing from SVI: build a strike grid and obtain IVs and call prices
# `band` is in moneyness K/S for convenience (e.g., (0.75, 1.30))
# ──────────────────────────────────────────────────────────────────────────────
def price_from_svi(out, band=(0.75, 1.30), nK=1201):
    S, r, q, T, F = out["S"], out["r"], out["q"], out["T"], out["F"]
    a, b, rho, m, sig = (out["params"][k] for k in ["a", "b", "rho", "m", "sigma"])

    # Build strike grid in K/S then to K
    K = np.linspace(band[0] * S, band[1] * S, nK)
    k = np.log(K / F)

    # Total variance -> IV
    w = np.maximum(svi_w(k, a, b, rho, m, sig), 1e-12)
    iv = np.sqrt(w / max(T, 1e-12))

    # Price calls from IVs
    C = bs_call_vec(S, K, T, r, q, iv)
    return K, iv, C


# ──────────────────────────────────────────────────────────────────────────────
# No-arbitrage checks in price space (discrete monotonicity and convexity)
# For calls: price must be decreasing and convex in K.
# ──────────────────────────────────────────────────────────────────────────────
def convexity_report(K, C):
    """
    Monotonicity: dC/dK <= 0 (violations counted where > +tol)
    Convexity:    d2C/dK2 >= 0 (violations counted where < -tol)
    """
    K = np.asarray(K, float)
    C = np.asarray(C, float)
    h = K[1] - K[0]
    dC = np.diff(C) / h
    d2C = np.diff(C, n=2) / (h * h)
    tol = 1e-10
    rep = {
        "min first diff": float(dC.min()),
        "max first diff": float(dC.max()),
        "min second diff": float(d2C.min()),
        "max second diff": float(d2C.max()),
        "monotonicity violations": int((dC > +tol).sum()),
        "convexity violations": int((d2C < -tol).sum()),
    }
    return rep


# ──────────────────────────────────────────────────────────────────────────────
# Breeden–Litzenberger density from a priced call grid
# q(K) = e^{rT} * d^2C/dK^2  (present-value convention)
# ──────────────────────────────────────────────────────────────────────────────
def bl_density_from_prices(K, C, r, T):
    K = np.asarray(K, float)
    C = np.asarray(C, float)
    h = K[1] - K[0]
    d2C = np.zeros_like(C)
    # central differences on interior points
    d2C[1:-1] = (C[:-2] - 2.0 * C[1:-1] + C[2:]) / (h * h)
    f = exp(r * T) * d2C
    return np.maximum(f, 0.0)  # clip tiny negatives due to discretization


def _trapz_subset(K, f, a, b):
    K = np.asarray(K, float)
    f = np.asarray(f, float)
    m = (K >= a) & (K <= b)
    if m.sum() < 2:
        return 0.0
    return float(np.trapz(f[m], K[m]))


# ──────────────────────────────────────────────────────────────────────────────
# Driver: run SVI over all days, apply dynamic evaluation band, compute diagnostics
# ──────────────────────────────────────────────────────────────────────────────
def run_svi_over_all_days(
    df_IV: pd.DataFrame,
    calib_band=(0.85, 1.15),   # select data with K/S in this band for fitting
    allowed_gap=0.10,          # extend eval band by ≤10% moneyness beyond observed per side
    eval_clamp=(0.75, 1.35),   # never evaluate outside these K/S clamps
    nK_eval=1201,
    robust=True,
    verbose=True,
):
    """
    For each unique quote_date in df_IV:
      • Fit SVI on a core calibration band (K/S)
      • Build a *dynamic* evaluation band = [mny_min_obs - allowed_gap, mny_max_obs + allowed_gap],
        clamped to eval_clamp
      • Price SVI on that band
      • Run no-arb checks and BL density, compute tail mass relative to observed coverage
      • Collect daily summary and artifacts
    """
    # Ensure expected columns and types
    needed = {"quote_date", "underlying_last", "risk_free_1m", "div_yield_annual", "dte", "strike", "implied_volatility"}
    missing_cols = needed - set(df_IV.columns)
    if missing_cols:
        raise KeyError(f"df_IV missing columns: {sorted(missing_cols)}")

    df = df_IV.copy()
    # Normalize date dtype
    df["quote_date"] = pd.to_datetime(df["quote_date"]).dt.date
    # Coerce numerics to be safe
    for c in ["underlying_last", "risk_free_1m", "div_yield_annual", "dte", "strike", "implied_volatility"]:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    dates = sorted(pd.Series(df["quote_date"]).dropna().unique())

    summary_rows = []
    artifacts = {}

    for d in dates:
        try:
            out = fit_svi(df, d, calib_band=calib_band, robust=robust)

            S, r, q, T = out["S"], out["r"], out["q"], out["T"]
            K_obs = out["K_obs"]
            mny_min_obs = float(np.min(K_obs) / S)
            mny_max_obs = float(np.max(K_obs) / S)

            # Dynamic evaluation band in K/S
            eval_min = max(mny_min_obs - allowed_gap, eval_clamp[0])
            eval_max = min(mny_max_obs + allowed_gap, eval_clamp[1])

            # Price SVI on the dynamic band
            K_grid, iv_grid, C_grid = price_from_svi(out, band=(eval_min, eval_max), nK=nK_eval)

            # No-arb report on the evaluation grid
            rep = convexity_report(K_grid, C_grid)

            # BL density and tail mass wrt observed coverage
            f = bl_density_from_prices(K_grid, C_grid, r, T)
            K_min_obs, K_max_obs = float(np.min(K_obs)), float(np.max(K_obs))

            total_eval_mass  = _trapz_subset(K_grid, f, K_grid[0], K_grid[-1])
            left_tail_mass   = _trapz_subset(K_grid, f, K_grid[0], min(K_min_obs, K_grid[-1]))
            right_tail_mass  = _trapz_subset(K_grid, f, max(K_max_obs, K_grid[0]), K_grid[-1])
            core_mass        = max(0.0, total_eval_mass - left_tail_mass - right_tail_mass)
            frac_outside_obs = (left_tail_mass + right_tail_mass) / total_eval_mass if total_eval_mass > 1e-10 else np.nan

            # Lee bounds from fitted params (info only; fit used soft constraints)
            lee_ok = bool(out["lee_ok"])
            slope_left = out["slope_left"]
            slope_right = out["slope_right"]

            summary_rows.append({
                "quote_date": d,
                "n_obs": int(len(K_obs)),
                "rmse_w": float(out["rmse_w"]),
                "mono_viol_inband": int(rep["monotonicity violations"]),
                "conv_viol_inband": int(rep["convexity violations"]),
                "lee_bound_violation": (not lee_ok),
                "slope_left": slope_left,
                "slope_right": slope_right,
                "mny_min_obs": mny_min_obs,
                "mny_max_obs": mny_max_obs,
                "eval_mny_min": eval_min,
                "eval_mny_max": eval_max,
                "extrap_left_mny": max(0.0, mny_min_obs - eval_min),
                "extrap_right_mny": max(0.0, eval_max - mny_max_obs),
                "extrap_max_mny": max(max(0.0, mny_min_obs - eval_min), max(0.0, eval_max - mny_max_obs)),
                "total_eval_mass": total_eval_mass,
                "left_tail_mass": left_tail_mass,
                "right_tail_mass": right_tail_mass,
                "core_mass": core_mass,
                "frac_outside_observed": frac_outside_obs,
                "res_status": int(out["res_status"]),
                "has_issue": (rep["monotonicity violations"] > 0) or (rep["convexity violations"] > 0) or (not lee_ok),
            })

            artifacts[d] = {
                "out": out,
                "K_grid": K_grid, "iv_grid": iv_grid, "C_grid": C_grid,
                "bl_density": f,
                "convexity_report": rep
            }

        except Exception as e:
            # If a day fails hard, record a stub so you know which one
            summary_rows.append({
                "quote_date": d,
                "n_obs": 0,
                "rmse_w": np.nan,
                "mono_viol_inband": 999,
                "conv_viol_inband": 999,
                "lee_bound_violation": True,
                "slope_left": np.nan,
                "slope_right": np.nan,
                "mny_min_obs": np.nan,
                "mny_max_obs": np.nan,
                "eval_mny_min": np.nan,
                "eval_mny_max": np.nan,
                "extrap_left_mny": np.nan,
                "extrap_right_mny": np.nan,
                "extrap_max_mny": np.nan,
                "total_eval_mass": np.nan,
                "left_tail_mass": np.nan,
                "right_tail_mass": np.nan,
                "core_mass": np.nan,
                "frac_outside_observed": np.nan,
                "res_status": 0,
                "has_issue": True,
                "notes": f"fit_error:{e}"
            })
            # No artifacts entry for failed day

    summary_df = pd.DataFrame(summary_rows).sort_values("quote_date").reset_index(drop=True)

    if verbose and not summary_df.empty:
        tot = len(summary_df)
        issues = int(summary_df["has_issue"].fillna(False).sum())
        print(f"Total days: {tot}")
        print(f"Days with in-band issues (mono/convex/Lee): {issues}")
        # Tail reliance summaries
        if np.isfinite(summary_df["frac_outside_observed"]).any():
            print("Median frac_outside_observed:", float(np.nanmedian(summary_df["frac_outside_observed"])))
        if np.isfinite(summary_df["total_eval_mass"]).any():
            print("Median total_eval_mass:", float(np.nanmedian(summary_df["total_eval_mass"])))

    return summary_df, artifacts


# ──────────────────────────────────────────────────────────────────────────────
# EXAMPLE USAGE (uncomment and run after you've prepared df_iv):
# ──────────────────────────────────────────────────────────────────────────────
# # 1) Ensure df_iv has expected column names; if your IV column is named differently, rename it:
df_iv = df_iv.rename(columns={"your_iv_col": "implied_volatility", "dte_days": "dte",
                              "risk_free_rate": "risk_free_1m", "Annualized_Dividend_Yield": "div_yield_annual"})
# #
# # 2) Run SVI across all days with dynamic evaluation band and tail diagnostics:
summary_df, artifacts = run_svi_over_all_days(
     df_iv,
     calib_band=(0.80, 1.15),   # fit on observed K/S in 0.85-1.15 (adjust to your taste)
     allowed_gap=0.10,          # ±10% moneyness beyond observed coverage for eval
     eval_clamp=(0.70, 1.35),   # never go outside these absolute clamps
     nK_eval=1201,
     robust=True,
     verbose=True
 )
display(summary_df.head(10))

# # 3) Inspect a specific day (example):
d0 = summary_df.loc[~summary_df["has_issue"], "quote_date"].iloc[0]
K, iv, C = artifacts[d0]["K_grid"], artifacts[d0]["iv_grid"], artifacts[d0]["C_grid"]
rep = artifacts[d0]["convexity_report"]
print("Convexity report:", rep)
```

```python
# ==============================
# BL pdf batch computation (no plots) — warning-free
# ==============================
import numpy as np
import pandas as pd
from math import exp

# ---------- Helpers ----------
def _second_derivative_nonuniform(x, y):
    r"""
    Compute the second derivative d²y/dx² on a 1D, monotone but possibly non-uniform grid.
    Uses np.gradient twice (with x as spacing) so non-uniform steps are handled.
    Endpoints use one-sided differences.
    Returns an array with the same shape as y.
    """
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    dy_dx   = np.gradient(y, x, edge_order=2)
    d2y_dx2 = np.gradient(dy_dx, x, edge_order=2)
    return d2y_dx2

def _bl_density_from_prices(K, C, r, T, floor_zero=True):
    r"""
    Breeden–Litzenberger density on the provided strike grid:
        q(K) = exp(r * T) * d²C/dK²
    Optionally floor negatives to zero to suppress numerical noise.
    """
    K = np.asarray(K, dtype=float)
    C = np.asarray(C, dtype=float)
    d2C = _second_derivative_nonuniform(K, C)
    q   = np.exp(r * T) * d2C
    if floor_zero:
        q = np.maximum(q, 0.0)
    return q

def _cumtrapz_grid(x, y):
    r"""
    Cumulative integral over the grid using the composite trapezoid rule.
    Returns F[i] = ∫_{x[0]}^{x[i]} y(u) du, with F[0] = 0.
    """
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    F = np.zeros_like(y, dtype=float)
    if len(x) < 2:
        return F
    dx = np.diff(x)
    area = 0.0
    for i in range(1, len(x)):
        area += 0.5 * (y[i-1] + y[i]) * dx[i-1]
        F[i] = area
    return F

def _trapz_subset(x, y, a, b):
    r"""
    Integral of y over [a, b] using trapezoids, clipping to [x[0], x[-1]].
    Ensures endpoints a and b are included via linear interpolation if needed.
    """
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    lo = max(a, x[0]); hi = min(b, x[-1])
    if hi <= lo:
        return 0.0
    mask = (x >= lo) & (x <= hi)
    xx, yy = x[mask], y[mask]
    # insert lo if missing
    if xx.size == 0 or xx[0] > lo:
        y_lo = np.interp(lo, x, y)
        xx   = np.r_[lo, xx]
        yy   = np.r_[y_lo, yy]
    # append hi if missing
    if xx[-1] < hi:
        y_hi = np.interp(hi, x, y)
        xx   = np.r_[xx, hi]
        yy   = np.r_[yy, y_hi]
    return float(np.trapz(yy, xx))

# ---------- Batch computation ----------
def compute_bl_pdfs_for_all_days(artifacts, summary_df=None, use_float32=False):
    r"""
    Compute BL PDFs for every day.

    Parameters
    ----------
    artifacts : dict
        Keyed by quote_date -> {
            "K_grid": array of strikes (evaluation grid),
            "C_grid": array of call prices aligned to K_grid,
            "iv_grid": array (unused here),
            "out": {"S": float, "r": float, "q": float, "T": float},
            "K_obs": array of traded strikes (optional; for tail/core accounting)
        }
    summary_df : DataFrame or None
        If provided and contains 'quote_date', controls processing order.
    use_float32 : bool
        If True, store arrays as float32 to save memory.

    Returns
    -------
    bl_results : dict
        date -> {
          "K_grid": K,
          "pdf": q(K),
          "cdf_eval": F_eval (integral on eval band; F_eval[-1] = total_eval_mass),
          "total_eval_mass": float,
          "left_tail_mass": float,
          "right_tail_mass": float,
          "core_mass": float,
          "S": float, "r": float, "q": float, "T": float
        }
    mass_summary_df : DataFrame
        Per-day mass diagnostics and eval/obs bounds.
    """
    # Determine processing order
    if summary_df is not None and "quote_date" in summary_df.columns:
        dates = list(summary_df["quote_date"])
    else:
        dates = list(artifacts.keys())
    dates = sorted(dates)

    bl_results = {}
    rows = []

    for d0 in dates:
        if d0 not in artifacts:
            continue

        art = artifacts[d0]
        out = art["out"]
        S, r, q_div, T = out["S"], out["r"], out["q"], out["T"]
        K_grid = np.array(art["K_grid"], dtype=float)
        C_grid = np.array(art["C_grid"], dtype=float)

        # 1) BL density on eval band
        f = _bl_density_from_prices(K_grid, C_grid, r, T, floor_zero=True)

        # 2) CDF (on eval band only)
        F_eval = _cumtrapz_grid(K_grid, f)
        total_eval_mass = float(F_eval[-1]) if F_eval.size else 0.0

        # 3) Tail/core mass accounting relative to observed strikes
        if "K_obs" in art and len(art["K_obs"]) > 0:
            K_obs = np.asarray(art["K_obs"], float)
            Kmin_obs, Kmax_obs = float(np.min(K_obs)), float(np.max(K_obs))
        else:
            # fall back to inner 5%–95% of eval band if no K_obs provided
            lo_idx = int(0.05 * (len(K_grid)-1)); hi_idx = int(0.95 * (len(K_grid)-1))
            Kmin_obs, Kmax_obs = K_grid[lo_idx], K_grid[hi_idx]

        left_tail_mass  = _trapz_subset(K_grid, f, K_grid[0], Kmin_obs)
        right_tail_mass = _trapz_subset(K_grid, f, Kmax_obs, K_grid[-1])
        core_mass       = max(0.0, total_eval_mass - left_tail_mass - right_tail_mass)

        # Optional memory compaction
        if use_float32:
            K_stored = K_grid.astype(np.float32)
            f_stored = f.astype(np.float32)
            F_stored = F_eval.astype(np.float32)
        else:
            K_stored, f_stored, F_stored = K_grid, f, F_eval

        bl_results[d0] = {
            "K_grid": K_stored,
            "pdf": f_stored,
            "cdf_eval": F_stored,
            "total_eval_mass": total_eval_mass,
            "left_tail_mass": left_tail_mass,
            "right_tail_mass": right_tail_mass,
            "core_mass": core_mass,
            "S": S, "r": r, "q": q_div, "T": T,
        }

        rows.append({
            "quote_date": d0,
            "total_eval_mass": total_eval_mass,
            "left_tail_mass": left_tail_mass,
            "right_tail_mass": right_tail_mass,
            "core_mass": core_mass,
            "Kmin_obs": Kmin_obs,
            "Kmax_obs": Kmax_obs,
            "Kmin_eval": float(K_grid[0]),
            "Kmax_eval": float(K_grid[-1]),
        })

    mass_summary_df = pd.DataFrame(rows).sort_values("quote_date")
    return bl_results, mass_summary_df

# --------- Run it ----------
# Requires: `artifacts` (dict as described) and optionally `summary_df` with quote_date.
bl_results, mass_summary = compute_bl_pdfs_for_all_days(
    artifacts,
    summary_df=summary_df,
    use_float32=True
)

# Quick sanity print
print("Computed BL pdfs for", len(bl_results), "days.")
print(mass_summary[["quote_date","total_eval_mass","left_tail_mass","right_tail_mass","core_mass"]].head(10))
```

```python
import numpy as np
import pandas as pd

def select_and_prepare_for_pit(bl_results, mass_summary, threshold, realized_prices):
    """
    Robust PIT prep:
      - filters mass_summary by threshold
      - aligns/normalizes dates across sources
      - augments CDF with tail masses
      - computes PIT values
    Returns a DataFrame with columns: quote_date, PIT_value, total_eval_mass
    """
    # ---- 1) Normalize dates everywhere ----
    # mass_summary
    ms = mass_summary.copy()
    ms["quote_date"] = pd.to_datetime(ms["quote_date"]).dt.normalize()

    # realized_prices
    rp = realized_prices.copy()
    if isinstance(rp, pd.Series):
        rp.index = pd.to_datetime(rp.index).normalize()
    elif isinstance(rp, dict):
        rp = pd.Series(rp)
        rp.index = pd.to_datetime(rp.index).normalize()
    else:
        # try to coerce generically
        rp = pd.Series(rp)
        rp.index = pd.to_datetime(rp.index).normalize()

    # bl_results as a dict keyed by date-like
    bl_map = {}
    for k, rec in bl_results.items():
        try:
            kd = pd.to_datetime(k).normalize()
        except Exception:
            # If keys are already datetimes
            kd = pd.to_datetime(pd.Timestamp(k)).normalize()
        bl_map[kd] = rec

    # ---- 2) Dates that pass threshold and exist in both bl_results and realized_prices ----
    passed = set(ms.loc[ms["total_eval_mass"] >= threshold, "quote_date"])
    have_bl = set(bl_map.keys())
    have_rp = set(rp.index)
    selected_dates = sorted(passed & have_bl & have_rp)

    if not selected_dates:
        # Return an empty but well-formed DataFrame
        return pd.DataFrame(columns=["quote_date", "PIT_value", "total_eval_mass"])

    # ---- 3) Compute PIT per date ----
    pit_records = []
    for d0 in selected_dates:
        rec = bl_map[d0]
        # Required arrays
        K_grid = np.asarray(rec.get("K_grid", []), dtype=float)
        F_eval = np.asarray(rec.get("cdf_eval", []), dtype=float)
        total_mass = float(rec.get("total_eval_mass", 0.0))
        left_mass  = float(rec.get("left_tail_mass", 0.0))
        right_mass = float(rec.get("right_tail_mass", 0.0))
        core_mass  = float(rec.get("core_mass", 0.0))

        # Skip if arrays missing or degenerate
        if K_grid.size < 2 or F_eval.size != K_grid.size or total_mass <= 0 or (left_mass + right_mass) >= 1.0:
            continue

        # Build augmented CDF on the eval band
        # F_eval starts at 0 and ends at total_mass (in-band). We want:
        #   F_aug(x) = left_mass + scale * F_eval(x), ending at 1 - right_mass
        scale = (1.0 - left_mass - right_mass) / core_mass if core_mass > 0 else 0.0
        if scale <= 0:
            # cannot sensibly scale; skip this day
            continue
        F_aug = left_mass + scale * F_eval
        # Numerical hygiene
        F_aug = np.clip(F_aug, 0.0, 1.0 - right_mass)
        F_aug[-1] = 1.0 - right_mass  # force exact end

        # Realized price
        y = float(rp.loc[d0])

        # PIT with tail handling
        if y <= K_grid[0]:
            pit_val = np.random.uniform(0.0, left_mass) if left_mass > 0 else 0.0
        elif y >= K_grid[-1]:
            lo = 1.0 - right_mass
            pit_val = np.random.uniform(lo, 1.0) if right_mass > 0 else 1.0
        else:
            pit_val = float(np.interp(y, K_grid, F_aug))

        pit_records.append({
            "quote_date": pd.to_datetime(d0),
            "PIT_value": pit_val,
            "total_eval_mass": total_mass
        })

    if not pit_records:
        return pd.DataFrame(columns=["quote_date", "PIT_value", "total_eval_mass"])

    df_pit = pd.DataFrame(pit_records).sort_values("quote_date")
    return df_pit
```

```python
def prepare_pit_df(bl_results, mass_summary, threshold, realized_prices):
    """
    Build PIT dataframe for all days that pass the eval mass threshold.
    """
    pit_records = []
    
    valid_days = mass_summary.loc[
        mass_summary["total_eval_mass"] >= threshold, "quote_date"
    ]

    for qd in valid_days:
        if qd not in bl_results or qd not in realized_prices:
            continue

        strikes = bl_results[qd]["K_grid"]
        pdf     = bl_results[qd]["pdf"]
        total_mass = bl_results[qd]["total_eval_mass"]

        # Normalize density to 1
        norm_factor = np.trapz(pdf, strikes)
        if norm_factor <= 0:
            continue
        pdf_norm = pdf / norm_factor

        # Realized underlying
        S_T = realized_prices[qd]

        # Compute PIT: ∫_{Kmin}^{S_T} f(K) dK
        mask = strikes <= S_T
        if not np.any(mask):
            pit_val = 0.0 if S_T < strikes.min() else 1.0
        else:
            pit_val = np.trapz(pdf_norm[mask], strikes[mask])

        pit_records.append({
            "quote_date": pd.to_datetime(qd),
            "realized": S_T,
            "PIT_value": pit_val,
            "total_eval_mass": total_mass
        })

    df_pit = pd.DataFrame(pit_records)
    if "quote_date" in df_pit.columns:
        df_pit = df_pit.sort_values("quote_date")
    return df_pit
```

```python
# ===============================================
# EVT (GPD) tail extension + PIT comparison cell
# ===============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import genpareto, kstest, ks_2samp

# ---------- GPD helper: span up to a tiny tail probability ----------
def _gpd_span_for_tol(xi, beta, tol=1e-6):
    """
    Return z_max such that CDF(z_max) = 1 - tol for a GPD(ξ, β), loc=0.
    Handles ξ=0 (exponential) and ξ>0.
    """
    tol = max(min(tol, 1-1e-12), 1e-12)  # clamp
    if abs(xi) < 1e-12:  # exponential
        return -beta * np.log(tol)
    else:
        # 1 - tol = 1 - (1 + xi z/beta)^(-1/xi)  ->  (1 + xi z/beta)^(-1/xi) = tol
        # (1 + xi z/beta) = tol^{-xi} -> z = beta/xi * (tol**(-xi) - 1)
        return beta / xi * (tol**(-xi) - 1.0)

# ---------- Build augmented (tail-extended) grid + pdf ----------
def augment_pdf_with_gpd(K, f, p_left, p_right, xi_L=0.0, xi_R=0.0, tail_prob_tol=1e-6):
    """
    Extend BL density with GPD tails that:
      - carry exactly p_left / p_right probability
      - are continuous with the in-band pdf at K_L / K_R (match pdf value at splice)
    Parameters:
      K, f         : in-band strike grid & density (numpy arrays)
      p_left/right : missing mass on each tail (from your bl_results)
      xi_L, xi_R   : GPD shape params (0 = exponential, >0 = heavier tails)
      tail_prob_tol: extend tails until remaining tail prob < tol
    Returns:
      K_aug, f_aug : augmented grid and normalized pdf
    """
    K = np.asarray(K, float)
    f = np.asarray(f, float)
    assert K.ndim == 1 and f.ndim == 1 and K.size == f.size and K.size >= 2

    K_L, K_R = K[0], K[-1]
    f_L, f_R = max(f[0], 1e-14), max(f[-1], 1e-14)  # avoid divide-by-0

    # Match continuity at splice: gpd.pdf(0; xi, beta) = 1/beta -> p_tail * (1/beta) = f_edge
    # => beta = p_tail / f_edge
    beta_L = p_left / f_L if p_left > 0 else 1.0
    beta_R = p_right / f_R if p_right > 0 else 1.0

    # Compute tail spans up to 'tail_prob_tol' residual probability
    zL_max = _gpd_span_for_tol(xi_L, beta_L, tol=tail_prob_tol) if p_left  > 0 else 0.0
    zR_max = _gpd_span_for_tol(xi_R, beta_R, tol=tail_prob_tol) if p_right > 0 else 0.0

    # Construct augmented grid: left tail, core, right tail
    n_core = len(K)
    n_tail = max(64, n_core // 4)  # reasonably smooth tails

    K_left  = (K_L - np.linspace(zL_max, 0.0, n_tail, endpoint=False)) if p_left  > 0 else np.array([])
    K_right = (K_R + np.linspace(0.0, zR_max, n_tail+1, endpoint=True))[1:] if p_right > 0 else np.array([])

    K_aug = np.concatenate([K_left, K, K_right])

    f_aug = np.zeros_like(K_aug)

    # Left tail density
    if p_left > 0:
        zL = K_L - K_left  # excesses ≥ 0
        f_left_shape = genpareto.pdf(zL, c=xi_L, loc=0.0, scale=beta_L)
        f_aug[:len(K_left)] = p_left * f_left_shape

    # Core: resample original f by interpolation to augmented grid portion
    core_slice = slice(len(K_left), len(K_left) + len(K))
    f_aug[core_slice] = np.interp(K_aug[core_slice], K, f)

    # Right tail density
    if p_right > 0:
        zR = K_right - K_R  # excesses ≥ 0
        f_right_shape = genpareto.pdf(zR, c=xi_R, loc=0.0, scale=beta_R)
        f_aug[-len(K_right):] = p_right * f_right_shape

    # Normalize exactly to 1
    area = np.trapz(f_aug, K_aug)
    if area > 0:
        f_aug /= area

    return K_aug, f_aug

# ---------- Compute PITs (baseline vs EVT-extended) ----------
def compute_pit(bl_results, mass_summary, realized_prices, threshold, use_evt=False, xi_L=0.0, xi_R=0.0, tail_prob_tol=1e-6):
    """
    Returns DataFrame with columns: quote_date, PIT_value, total_eval_mass
    Always returns these columns even if empty (no KeyError later).
    """
    # Normalize date types
    mass_df = mass_summary.copy()
    mass_df["quote_date"] = pd.to_datetime(mass_df["quote_date"])
    bl_map = {pd.to_datetime(k): v for k, v in bl_results.items()}
    rp_map = {pd.to_datetime(k): v for k, v in (realized_prices.items() if hasattr(realized_prices, 'items') else dict(realized_prices).items())}

    rows = []
    valid_dates = mass_df.loc[mass_df["total_eval_mass"] >= threshold, "quote_date"]

    for d in valid_dates:
        if d not in bl_map or d not in rp_map:
            continue
        rec = bl_map[d]
        K = rec["K_grid"]; f = rec["pdf"]
        pL = float(rec.get("left_tail_mass", 0.0))
        pR = float(rec.get("right_tail_mass", 0.0))
        S_real = float(rp_map[d])

        if use_evt:
            K_use, f_use = augment_pdf_with_gpd(K, f, pL, pR, xi_L=xi_L, xi_R=xi_R, tail_prob_tol=tail_prob_tol)
        else:
            # Baseline: normalize in-band pdf to 1 (so PIT lives on [0,1] cleanly)
            area = np.trapz(f, K)
            if area <= 0:
                continue
            K_use, f_use = K, f / area

        # Build CDF
        F = np.zeros_like(f_use)
        F[1:] = np.cumsum(0.5 * (f_use[1:] + f_use[:-1]) * np.diff(K_use))
        if F[-1] > 0:
            F = F / F[-1]

        # PIT value at realized price
        pit = float(np.interp(S_real, K_use, F))
        rows.append({"quote_date": d, "PIT_value": pit, "total_eval_mass": rec.get("total_eval_mass", np.nan)})

    # Guarantee columns exist even if empty
    if not rows:
        return pd.DataFrame({"quote_date": [], "PIT_value": [], "total_eval_mass": []})

    return pd.DataFrame(rows).sort_values("quote_date").reset_index(drop=True)

# ---------- USER CONTROLS ----------
thr             = 0.998   # threshold for total_eval_mass
xi_L, xi_R      = 0.2, 0.20   # tail shape (0 = exponential; >0 = heavier)
tail_prob_tol   = 1e-8    # how far tails extend (probability left beyond far endpoint)

# ---------- Run ----------
df_pit_base = compute_pit(bl_results, mass_summary, realized_prices, threshold=thr, use_evt=False)
df_pit_evt  = compute_pit(bl_results, mass_summary, realized_prices, threshold=thr, use_evt=True,
                          xi_L=xi_L, xi_R=xi_R, tail_prob_tol=tail_prob_tol)

print("Baseline PIT rows:", len(df_pit_base))
print("EVT-extended PIT rows:", len(df_pit_evt))

# ---------- KS tests vs Uniform(0,1) ----------
def ks_uniform(df, label):
    if df.empty:
        print(f"{label}: empty — skipping KS vs Uniform")
        return None, None
    stat, p = kstest(df["PIT_value"].values, 'uniform')
    print(f"{label}: KS vs Uniform -> stat={stat:.4f}, p={p:.3e}")
    return stat, p

ks_uniform(df_pit_base, "Baseline")
ks_uniform(df_pit_evt,  "EVT-extended")

# ---------- Two-sample KS between PIT sets ----------
if not df_pit_base.empty and not df_pit_evt.empty:
    stat2, p2 = ks_2samp(df_pit_base["PIT_value"].values, df_pit_evt["PIT_value"].values)
    print(f"Baseline vs EVT PITs -> KS stat={stat2:.4f}, p={p2:.3e}")
else:
    print("Two-sample KS: skipped (one set empty)")

# ---------- Plots ----------
fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)
if not df_pit_base.empty:
    axes[0].hist(df_pit_base["PIT_value"], bins=12, edgecolor="k", alpha=0.75)
else:
    axes[0].text(0.5, 0.5, "No data", ha="center", va="center", transform=axes[0].transAxes)
axes[0].set_title(f"Baseline PIT (thr={thr})")
axes[0].set_xlabel("PIT value"); axes[0].set_ylabel("Count")

if not df_pit_evt.empty:
    axes[1].hist(df_pit_evt["PIT_value"], bins=12, edgecolor="k", alpha=0.75, color="orange")
else:
    axes[1].text(0.5, 0.5, "No data", ha="center", va="center", transform=axes[1].transAxes)
axes[1].set_title(f"EVT-extended PIT (thr={thr}, xi_L={xi_L}, xi_R={xi_R}, tol={tail_prob_tol})")
axes[1].set_xlabel("PIT value")

plt.tight_layout()
plt.show()
```

## Using Breeden–Litzenberger PDFs as a Binary Classifier

### 1. From Risk-Neutral PDF to Up/Down Probabilities
The Breeden–Litzenberger method yields a **risk-neutral probability density function (pdf)** \(q(K)\) over strikes \(K\).  
From this we can compute **cumulative probabilities** of moving above or below the current forward level \(F\):

$
P^{Q}_{\text{up}} = \int_{F}^{\infty} q(K)\, dK, 
\qquad 
P^{Q}_{\text{down}} = \int_{0}^{F} q(K)\, dK
$

- $(P^{Q}_{\text{up}}$: risk-neutral probability of finishing above the forward.  
- $(P^{Q}_{\text{down}}$: risk-neutral probability of finishing below the forward.  

Note: these are **not physical probabilities** — they are distorted by risk premia — but they represent the market’s *pricing-implied weights* on up vs down moves.

---

### 2. Defining the Prediction Task
We can reframe the problem as a **binary classification**:

- **Target (label):** realized outcome
  - `1` if $(S_T > S_0$ (or forward $(F)$
  - `0` otherwise  

- **Features (predictors):**
  - $P^{Q}_{\text{up}}$ and $P^{Q}_{\text{down}}$  
  - Or just one of them, since they sum to ~1  
  - Possibly enriched with diagnostics (e.g., implied volatility, tail mass, skewness from BL pdf)

---

### 3. Classifier Approaches
- **Naïve threshold rule:**  
  Predict `up` if \(P^{Q}_{\text{up}} > 0.5\), else `down`.  

- **Logistic regression / ML:**  
  Use \(P^{Q}_{\text{up}}\) as input into a logistic model to estimate how well it explains realized outcomes.  
  This allows testing whether BL probabilities are systematically biased vs reality.  

---

### 4. Evaluation
We evaluate predictive power using classification metrics:
- **Accuracy:** fraction of correct up/down predictions.  
- **AUC (ROC):** sensitivity to thresholds.  
- **Calibration plots:** compare predicted \(P^{Q}_{\text{up}}\) vs empirical frequency of up moves.  

---

### 5. Interpretation
- If risk-neutral pdfs have no directional power, accuracy ≈ 50% and calibration is poor.  
- If there is some link, we might see **systematic skew** (e.g., RN overstates downside risk).  
- Either way, results provide an empirical test of whether option-implied distributions contain *any* information about direction under the physical measure.


```python
# ========================================
# Naive BL-based Direction Classifier (robust)
# ========================================
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

def _coerce_realized_to_df(realized_prices_like):
    """
    Accepts either:
      - pd.Series with DatetimeIndex (name can be 'quote_date' or None), or
      - pd.DataFrame with columns ['quote_date', 'realized_price'] (any order),
    and returns a clean DataFrame with normalized datetime column 'quote_date'
    and float column 'realized_price'.
    """
    if isinstance(realized_prices_like, pd.Series):
        df = realized_prices_like.copy()
        # If the index is named something else, we reset and rename explicitly
        df = df.reset_index()
        if df.shape[1] != 2:
            raise ValueError("Realized Series must become a 2-column DataFrame after reset_index().")
        df.columns = ["quote_date", "realized_price"]
    elif isinstance(realized_prices_like, pd.DataFrame):
        df = realized_prices_like.copy()
        # Try to find the date and price columns
        cols = {c.lower(): c for c in df.columns}
        date_col = cols.get("quote_date", None)
        price_col = cols.get("realized_price", None)
        if date_col is None:
            # if date col missing, assume first col is date
            date_col = df.columns[0]
        if price_col is None:
            # if price col missing, assume last col is realized price
            price_col = df.columns[-1]
        df = df[[date_col, price_col]].rename(columns={date_col: "quote_date", price_col: "realized_price"})
    else:
        # Try dict-like
        try:
            ser = pd.Series(realized_prices_like)
            df = ser.reset_index()
            df.columns = ["quote_date", "realized_price"]
        except Exception as e:
            raise ValueError("Unsupported realized_prices type; pass a Series or DataFrame.") from e

    # Normalize types
    df["quote_date"] = pd.to_datetime(df["quote_date"]).dt.normalize()
    df["realized_price"] = pd.to_numeric(df["realized_price"], errors="coerce")
    df = df.dropna(subset=["quote_date", "realized_price"]).reset_index(drop=True)
    return df

def _build_bl_df(bl_results):
    """
    Turn your bl_results dict into a tidy DataFrame with one row per date:
    columns: quote_date, K_grid, pdf, S, r, q, T, total_eval_mass
    """
    rows = []
    for k, rec in bl_results.items():
        d = pd.to_datetime(k).normalize()
        rows.append({
            "quote_date": d,
            "K_grid": rec["K_grid"],
            "pdf": rec["pdf"],
            "S": float(rec["S"]),
            "r": float(rec["r"]),
            "q": float(rec["q"]),
            "T": float(rec["T"]),
            "total_eval_mass": float(rec.get("total_eval_mass", np.nan)),
        })
    df = pd.DataFrame(rows).sort_values("quote_date").reset_index(drop=True)
    return df

def _rn_up_probability(K, f, F):
    """
    Risk-neutral P_up = ∫_{K>=F} f(K) dK, after normalizing f to integrate to 1 over its grid.
    """
    K = np.asarray(K, float)
    f = np.asarray(f, float)
    # Normalize by actual area on the provided grid
    area = np.trapz(f, K)
    if area <= 0:
        return np.nan
    f = f / area
    mask = K >= F
    if not np.any(mask):
        return 0.0
    return float(np.trapz(f[mask], K[mask]))

def naive_bl_direction_classifier(bl_results, realized_prices_like, threshold=0.5, direction_vs="forward"):
    """
    Naive binary classifier using BL pdf:
      - Compute RN up-probability P_up from BL (vs forward F).
      - Predict UP if P_up > threshold, else DOWN.
      - Realized direction: compare realized_price to either spot S or forward F (choose via direction_vs).
      - Return per-day results and overall accuracy + confusion matrix.

    Params
    ------
    bl_results : dict from compute_bl_pdfs_for_all_days (keys are dates)
    realized_prices_like : Series or DataFrame of realized prices
    threshold : float, default 0.5
    direction_vs : {"forward","spot"}
        If "forward": realized_direction = 1{ realized_price > F }
        If "spot"   : realized_direction = 1{ realized_price > S }

    Returns
    -------
    results_df : DataFrame with columns
        quote_date, realized_price, S, F, P_up, predicted_label, realized_label, correct
    metrics : dict with accuracy and confusion matrix (2x2 ndarray with labels [DOWN, UP])
    """
    # Coerce inputs to tidy frames
    df_real = _coerce_realized_to_df(realized_prices_like)
    df_bl   = _build_bl_df(bl_results)

    # Merge on quote_date
    df = df_bl.merge(df_real, on="quote_date", how="inner")
    if df.empty:
        # Return empty but well-formed outputs
        return pd.DataFrame(columns=["quote_date","realized_price","S","F","P_up","predicted_label","realized_label","correct"]), {"accuracy": np.nan, "confusion_matrix": np.array([[0,0],[0,0]])}

    # Compute forward, P_up, and labels
    F_vals = df["S"] * np.exp((df["r"] - df["q"]) * df["T"])
    df["F"] = F_vals

    P_up_list = []
    for K, f, F in zip(df["K_grid"], df["pdf"], df["F"]):
        P_up_list.append(_rn_up_probability(K, f, F))
    df["P_up"] = P_up_list

    # Prediction
    df["predicted_label"] = (df["P_up"] > threshold).astype(int)

    # Realized direction vs chosen benchmark
    if direction_vs == "spot":
        df["realized_label"] = (df["realized_price"] > df["S"]).astype(int)
    else:  # "forward"
        df["realized_label"] = (df["realized_price"] > df["F"]).astype(int)

    # Correct?
    df["correct"] = (df["predicted_label"] == df["realized_label"]).astype(int)

    # Metrics
    if len(df) > 0:
        acc = df["correct"].mean()
        # confusion matrix with labels [DOWN=0, UP=1]
        cm = confusion_matrix(df["realized_label"], df["predicted_label"], labels=[0,1])
    else:
        acc = np.nan
        cm = np.array([[0,0],[0,0]])

    results_df = df[["quote_date","realized_price","S","F","P_up","predicted_label","realized_label","correct"]].sort_values("quote_date").reset_index(drop=True)
    metrics = {"accuracy": float(acc), "confusion_matrix": cm}
    return results_df, metrics

# =======================
# Run it
# =======================
# If your realized_prices is still a Series:
# realized_prices_df = realized_prices.reset_index().rename(columns={"index":"quote_date", 0:"realized_price"})
# But we can pass the Series directly; the function will coerce it.

results_df, metrics = naive_bl_direction_classifier(
    bl_results=bl_results,
    realized_prices_like=realized_prices,   # Series or DataFrame accepted
    threshold=0.5,
    direction_vs="forward"  # or "spot,forward"
)


print(f"Naive BL direction accuracy: {metrics['accuracy']:.3f}  (N={len(results_df)})")
print("Confusion matrix [rows=actual DOWN/UP, cols=predicted DOWN/UP]:\n", metrics["confusion_matrix"])
print(results_df.head())
```
